## 1. Setup and Data Loading

This section prepares the analysis environment and loads the HealthBench dataset. The dataset contains real medical conversations that have been carefully evaluated by healthcare professionals using detailed scoring rubrics.

**What We're Loading**: A collection of 5,000 medical conversations, each with:
- Patient questions or concerns
- AI system responses
- Professional evaluation scores across multiple criteria
- Medical category classifications
- Conversation metadata


```python
# Import all necessary libraries
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import re
from collections import Counter
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from wordcloud import WordCloud

# Set plot style
plt.style.use('seaborn-v0_8-whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Load the JSONL dataset
records = []
with open('2025-05-07-06-14-12_oss_eval.jsonl', 'r') as f:
    for line in f:
        records.append(json.loads(line))
df = pd.json_normalize(records)
print(f"Loaded {len(df)} examples.")

# Display dataset structure
print("\nDataset columns:")
print(df.columns.tolist())

# Basic info about the dataset
print("\nSample record structure:")
sample_record = records[0]
for key, value in sample_record.items():
    if isinstance(value, list) and len(value) > 0:
        print(f"{key}: List with {len(value)} items")
    else:
        print(f"{key}: {type(value).__name__}")
```

## 2. Rubric Analysis

Medical professionals evaluate AI responses using detailed rubrics - structured scoring systems that assess different aspects of medical communication. This section analyzes how these evaluation criteria are distributed and applied.

**Understanding Rubrics**: Each conversation is scored across multiple criteria such as:
- Medical accuracy and completeness
- Communication clarity and empathy
- Safety considerations and appropriate referrals
- Cultural sensitivity and patient-centered care

**Why This Matters**: Understanding rubric patterns helps identify what medical professionals consider most important when evaluating AI medical advice, and reveals the complexity of medical communication assessment.


### 2.1 Distribution of Rubric Counts

**What This Shows**: How many evaluation criteria are applied to each medical conversation. Some conversations may be simple and require fewer evaluation points, while complex medical discussions need more comprehensive assessment.

**Clinical Relevance**: The number of rubric criteria often reflects the complexity of the medical scenario. More criteria typically indicate conversations involving multiple medical considerations, safety concerns, or nuanced clinical decision-making.


```python
# Calculate number of rubrics per example
df['num_rubrics'] = df['rubrics'].apply(lambda x: len(x) if isinstance(x, list) else 0)

# Plot distribution
plt.hist(df['num_rubrics'], bins=range(df['num_rubrics'].min(), df['num_rubrics'].max()+2),
         edgecolor='black', color='skyblue')
plt.xlabel('Number of Rubric Criteria')
plt.ylabel('Number of Examples')
plt.title('Rubric Counts per Example')

# Add summary statistics
print(f"Rubric count statistics:")
print(f"Mean: {df['num_rubrics'].mean():.2f}")
print(f"Median: {df['num_rubrics'].median():.2f}")
print(f"Min: {df['num_rubrics'].min()}")
print(f"Max: {df['num_rubrics'].max()}")
plt.show()
```

### 2.2 Rubric Tag Analysis

**What This Shows**: The most frequently used evaluation categories across all medical conversations. Tags represent specific aspects of medical communication that evaluators consistently assess.

**Understanding the Tags**: Common tags might include categories like "medical-accuracy", "empathy", "safety", or "cultural-competence". The frequency of these tags reveals what medical professionals prioritize when evaluating AI medical responses.

**Clinical Insight**: High-frequency tags indicate the core competencies expected in medical AI systems, while rare tags may represent specialized evaluation criteria for specific medical scenarios.


```python
# Extract all tags from all rubrics
all_tags = df['rubrics'].explode().dropna().apply(lambda r: r.get('tags', [])).explode()
tag_counts = all_tags.value_counts().head(10)

# Plot top 10 tags
tag_counts.plot.bar(color='lightgreen')
plt.xlabel('Rubric Tag')
plt.ylabel('Frequency')
plt.title('Top 10 Rubric Tags')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Display tag statistics
print(f"Total unique tags: {all_tags.nunique()}")
print(f"\nTop 10 tags:")
display(tag_counts)
```

### 2.3 Tag Co-occurrence Analysis

**What This Shows**: Which evaluation criteria are commonly assessed together in the same conversation. This reveals patterns in how medical professionals structure their comprehensive evaluations.

**Clinical Significance**: Certain evaluation criteria naturally go together. For example, conversations requiring safety assessment might also need empathy evaluation, or discussions about treatment options might require both accuracy and cultural competence assessment.

**Practical Application**: Understanding these patterns helps identify comprehensive evaluation frameworks and ensures that related aspects of medical communication are consistently assessed together.


```python
def get_tag_pairs(tags):
    """Extract all unique pairs of tags from a list of tags"""
    pairs = []
    if isinstance(tags, list) and len(tags) > 1:
        for i in range(len(tags)):
            for j in range(i+1, len(tags)):
                pairs.append(tuple(sorted([tags[i], tags[j]])))
    return pairs

# Extract all tag pairs from rubrics
all_tag_pairs = []
for rubrics in df['rubrics']:
    if isinstance(rubrics, list):
        for rubric in rubrics:
            all_tag_pairs.extend(get_tag_pairs(rubric.get('tags', [])))

# Count and display top co-occurring pairs
pair_counts = pd.Series(all_tag_pairs).value_counts().head(10)
plt.figure(figsize=(12, 6))
pair_counts.plot(kind='bar', color='coral')
plt.title('Top 10 Co-occurring Tag Pairs')
plt.xlabel('Tag Pairs')
plt.ylabel('Co-occurrence Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print('\nTop 10 co-occurring tag pairs:')
display(pair_counts)
```

### 2.4 Rubric Point Value Analysis

---



**What This Shows**: The distribution of point values assigned to different evaluation criteria. Point values indicate the relative importance or weight given to different aspects of medical communication.

**Understanding Point Values**: Higher point values typically indicate more critical evaluation criteria. For example, medical accuracy might receive more points than communication style, reflecting the paramount importance of correct medical information.

**Clinical Interpretation**: The point distribution reveals the evaluation priorities of medical professionals and helps understand which aspects of AI medical responses are considered most crucial for patient safety and care quality.


```python
# Create a flattened dataframe of all rubrics
all_rubrics = pd.json_normalize(df['rubrics'].explode().dropna())

# Plot distribution of point values
plt.hist(all_rubrics['points'], bins=sorted(set(all_rubrics['points'])),
         edgecolor='black', color='lightblue')
plt.xlabel('Point Value')
plt.ylabel('Number of Criteria')
plt.title('Rubric Point Value Distribution')
plt.show()

# Display point value statistics
print(f"Point value statistics:")
print(f"Mean: {all_rubrics['points'].mean():.2f}")
print(f"Median: {all_rubrics['points'].median():.2f}")
print(f"Min: {all_rubrics['points'].min()}")
print(f"Max: {all_rubrics['points'].max()}")
print(f"\nPoint value frequency:")
display(all_rubrics['points'].value_counts().sort_index())
```

### 2.5 Rubric Axis Analysis

**What This Shows**: How evaluation criteria are organized into major assessment categories or "axes". These axes represent fundamental dimensions of medical communication quality.

**Understanding Axes**: Common axes might include:
- **Clinical Competence**: Medical knowledge and diagnostic accuracy
- **Communication Skills**: Clarity, empathy, and patient engagement
- **Safety and Ethics**: Risk assessment and ethical considerations
- **Cultural Competence**: Sensitivity to diverse patient backgrounds

**Clinical Value**: This analysis reveals how medical professionals structure comprehensive evaluations and the relative emphasis placed on different fundamental aspects of medical care.


```python
# Extract axis from tags
def extract_axis(tags):
    """Extract the axis value from a list of tags"""
    for tag in tags:
        if tag.startswith('axis:'):
            return tag.split(':', 1)[1]
    return None

# Add axis to the rubrics dataframe
all_rubrics['axis'] = all_rubrics['tags'].apply(extract_axis)

# Summarize by axis
axis_summary = all_rubrics.dropna(subset=['axis']).groupby('axis')['points'].agg(
    num_criteria='count',
    avg_points='mean',
    min_points='min',
    max_points='max'
).round(2).reset_index()

print('Rubric Axis Summary:')
display(axis_summary)

# Visualize point distribution by axis
plt.figure(figsize=(12, 6))
all_rubrics.boxplot(column='points', by='axis', figsize=(12, 6))
plt.title('Point Distribution by Axis')
plt.suptitle('')  # Remove automatic suptitle
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 2.6 Relationship between Rubric Complexity and Points

**What This Shows**: Whether conversations with more evaluation criteria also tend to have higher total point values, and how strongly these factors are related.

**Clinical Interpretation**: A strong correlation suggests that complex medical scenarios (requiring many evaluation criteria) also tend to be high-stakes situations (with many points at stake). This relationship helps understand the connection between conversation complexity and evaluation importance.

**Practical Implications**: Understanding this relationship helps predict evaluation workload and identify conversations that may require more detailed professional review due to their complexity and importance.


```python
# Calculate complexity metrics
df['total_points'] = df['rubrics'].apply(lambda x: sum(r['points'] for r in x) if isinstance(x, list) else 0)
df['avg_points_per_rubric'] = df['total_points'] / df['num_rubrics'].replace(0, float('nan'))

# Create scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['num_rubrics'], df['total_points'], alpha=0.5, color='purple')
plt.xlabel('Number of Rubric Criteria')
plt.ylabel('Total Points')
plt.title('Relationship between Rubric Complexity and Total Points')

# Add trend line
z = np.polyfit(df['num_rubrics'], df['total_points'], 1)
p = np.poly1d(z)
plt.plot(df['num_rubrics'], p(df['num_rubrics']), 'r--', alpha=0.8,
         label=f'y={z[0]:.2f}x+{z[1]:.2f}')
plt.legend()
plt.show()

# Display correlation and summary statistics
correlation = df['num_rubrics'].corr(df['total_points'])
print(f"Correlation between number of rubrics and total points: {correlation:.3f}")
print(f"Average points per rubric: {df['avg_points_per_rubric'].mean():.2f}")
```

## 3. Medical Category and Theme Analysis

This section examines the medical specialties and themes represented in the conversations. Understanding the distribution of medical topics helps identify which areas of medicine are most commonly addressed by AI systems and patients.

**Clinical Relevance**: Different medical specialties may require different communication approaches and evaluation criteria. For example, mental health conversations might emphasize empathy and safety, while diagnostic discussions might prioritize accuracy and thoroughness.

**Dataset Balance**: Analyzing category distribution helps understand whether the dataset provides balanced representation across medical specialties, which is crucial for developing AI systems that work well across all areas of medicine.


```python
# Extract medical categories and themes
df['medical_category'] = df['example_tags'].apply(
    lambda x: next((tag.replace('physician_agreed_category:', '')
                    for tag in x if 'physician_agreed_category:' in tag), None))
df['theme'] = df['example_tags'].apply(
    lambda x: next((tag.replace('theme:', '') for tag in x if 'theme:' in tag), None))

# Medical Category Analysis
print('Medical Category Distribution:')
med_cat_dist = df['medical_category'].value_counts()
display(med_cat_dist)

plt.figure(figsize=(12, 6))
med_cat_dist.plot(kind='bar', color='lightseagreen')
plt.title('Distribution of Medical Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Theme Analysis
print('\nTheme Distribution:')
theme_dist = df['theme'].value_counts().head(15)
display(theme_dist)

plt.figure(figsize=(12, 6))
theme_dist.plot(kind='bar', color='sandybrown')
plt.title('Top 15 Themes')
plt.xlabel('Theme')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## 4. Conversation Analysis

This section analyzes the structure and characteristics of medical conversations, including dialogue patterns, message lengths, and communication styles. Understanding these patterns provides insights into how patients and AI systems interact in medical contexts.

**Why Conversation Structure Matters**: The way medical conversations unfold affects both patient satisfaction and clinical outcomes. Single-turn conversations might indicate straightforward information requests, while multi-turn dialogues suggest more complex medical discussions requiring clarification and follow-up.

**Communication Patterns**: Analyzing conversation characteristics helps identify effective communication strategies and reveals differences between patient language and clinical language that can inform AI system design.


### 4.1 Turn Analysis

**What This Shows**: The distribution of conversation lengths, measured by the number of back-and-forth exchanges between patients and the AI system.

**Clinical Significance**:
- **Single-turn conversations** typically represent direct questions with straightforward answers (e.g., "What are the side effects of aspirin?")
- **Multi-turn conversations** suggest more complex medical discussions requiring clarification, follow-up questions, or detailed explanations

**Patient Care Implications**: Understanding turn patterns helps identify when patients need more extensive dialogue and support, versus when they can be satisfied with direct, concise responses.


```python
# Calculate conversation metrics
df['turns'] = df['prompt'].apply(len)
df['first_user_message'] = df['prompt'].apply(lambda p: next((turn['content'] for turn in p if turn['role'] == 'user'), ''))
df['message_len'] = df['first_user_message'].apply(len)

# Single vs multi-turn analysis
single_turn = df[df['turns'] == 1].shape[0]
multi_turn = df[df['turns'] > 1].shape[0]
print(f"Single-turn conversations: {single_turn} ({single_turn/len(df)*100:.2f}%)")
print(f"Multi-turn conversations: {multi_turn} ({multi_turn/len(df)*100:.2f}%)")

# Turn distribution plot
plt.figure(figsize=(10, 6))
turn_counts = df['turns'].value_counts().sort_index()
plt.bar(turn_counts.index, turn_counts.values, color='steelblue')
plt.title('Distribution of Conversation Turn Counts')
plt.xlabel('Number of Turns')
plt.ylabel('Count')
plt.xticks(range(1, df['turns'].max() + 1))
plt.show()
```

### 4.2 Message Length Analysis

**Clinical Insights**:

### 4.3 Content Analysis by Role

This analysis examines the most frequent words used by patients vs. clinicians. We filter out:
- Single letter words (often artifacts or abbreviations without context)
- Numbers (dates, measurements that don't indicate semantic content)
- Additional common words in multiple languages (de, la, etc.) that don't provide meaningful insights
- Standard English stopwords

This filtering helps reveal the substantive vocabulary differences between patient language and clinical language.


```python
# Prepare containers for text by role
user_text = []
assistant_text = []

# Extract text by role
for _, row in df.iterrows():
    for msg in row['prompt']:
        if msg['role'] == 'user':
            user_text.append(msg['content'])
        elif msg['role'] == 'assistant':
            assistant_text.append(msg['content'])

# Define tokenization and enhanced stopwords
def tokenize(text):
    return re.findall(r'\b\w+\b', text.lower())

# Enhanced stopwords including single letters, numbers, and multilingual common words
stopwords = set([
    # Standard English stopwords
    'the', 'and', 'to', 'a', 'of', 'i', 'in', 'is', 'it', 'for', 'on', 'my', 'with', 'that', 'you', 'have',
    'are', 'be', 'as', 'was', 'but', 'at', 'or', 'so', 'if', 'can', 'do', 'me', 'am', 'not', 'this', 'your',
    'we', 'by', 'from', 'an', 'has', 'had', 'will', 'would', 'should', 'could', 'what', 'when', 'which', 'how',
    'about', 'just', 'like', 'get', 'got', 'been', 'they', 'he', 'she', 'him', 'her', 'their', 'them', 'our',
    'us', 'who', 'all', 'any', 'some', 'no', 'yes', 'more', 'also', 'than', 'out', 'up', 'down', 'into', 'over',
    'under', 'after', 'before', 'because', 'why', 'where', 'there', 'then', 'now', 'only', 'too', 'very',
    # Additional multilingual and common words
    'de', 'la', 'el', 'en', 'un', 'una', 'le', 'les', 'du', 'des', 'et', 'ou', 'ne', 'se', 'ce', 'sa', 're', 'son',
    'sur', 'avec', 'pour', 'par', 'dans', 'sans', 'sous', 'entre', 'depuis', 'pendant', 'avant', 'après',
    # Single letters (common artifacts)
    'a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j', 'k', 'l', 'm', 'n', 'o', 'p', 'q', 'r', 's', 't', 'u', 'v', 'w', 'x', 'y', 'z'
])

def get_top_words(texts, stopwords, topn=30):
    words = []
    for text in texts:
        words.extend(tokenize(text))
    # Filter out stopwords, single letters, and numbers
    words = [w for w in words if w not in stopwords and len(w) > 1 and not w.isdigit()]
    word_counts = Counter(words)
    return word_counts.most_common(topn)

# Get top 30 words for each role
user_top30 = get_top_words(user_text, stopwords, 30)
assistant_top30 = get_top_words(assistant_text, stopwords, 30)

# Plot function for top 10 visualization
def plot_bar_and_wordcloud(top_words, title_prefix):
    # Use top 10 for visualization
    top10 = top_words[:10]
    words_, counts_ = zip(*top10)
    plt.figure(figsize=(10,5))
    plt.bar(words_, counts_, color='skyblue')
    plt.title(f'{title_prefix}: Top 10 Words')
    plt.ylabel('Frequency')
    plt.xlabel('Word')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(top10))
    plt.figure(figsize=(10,5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud: {title_prefix}')
    plt.show()

# Plot for user (patient)
plot_bar_and_wordcloud(user_top30, "Patient Prompts")

# Display top 30 words as comma-separated list
user_words = [word for word, count in user_top30]
print("Top 30 Patient Words:")
print(", ".join(user_words))

# Plot for assistant (clinician)
plot_bar_and_wordcloud(assistant_top30, "Clinician Responses")

# Display top 30 words as comma-separated list
assistant_words = [word for word, count in assistant_top30]
print("\nTop 30 Clinician Words:")
print(", ".join(assistant_words))
```
